```python
import pandas as pd
import numpy as np
data=pd.read_excel("BCLS.xlsx",sheet_name="Sheet1")
data
```

```python
i=data.columns
Class=list(data[i[5]])
p_y=(Class.count('yes')/14)
p_n=(Class.count('no')/14)
py=[]
pn=[]
c_n,c_y = [x for _, x in data.groupby(data[i[5]] == 'yes')]
```

```python
k=1
for j in ['youth','medium','yes','fair']:
    py.append(list(c_y[i[k]]).count(j)/9)
    pn.append(list(c_n[i[k]]).count(j)/5)
    k+=1
print(py)
print('\n')
print(pn)
py.append(p_y)
pn.append(p_n)
print('\n\n')
t_y=t_n=1
for m,n in zip(py,pn):
    t_y*=m
    t_n*=n
if t_y > t_n:
    print("Class: Yes")
else:
    print("Class: No")
```
